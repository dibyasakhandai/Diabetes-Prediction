Training the data

```python
X = diabetes_dataset[['Glucose','BMI','Age','Insulin','SkinThickness']]
Y = diabetes_dataset['Outcome']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

scalar = StandardScaler()
scalar.fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)

classifier = svm.SVC(kernel='linear', probability=True)
classifier.fit(X_train, Y_train)

import joblib
joblib.dump(classifier, "diabetes_model.pkl")
joblib.dump(scalar, "scaler.pkl")
```

Model Evalution

Accuracy  Score

```python
#accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib

# Load model and scaler
classifier = joblib.load("diabetes_model.pkl")
scalar = joblib.load("scaler.pkl")

# Prediction function
def predict_diabetes(data):
    input_df = pd.DataFrame([data])
    input_scaled = scalar.transform(input_df)
    pred_class = classifier.predict(input_scaled)[0]
    pred_prob = classifier.decision_function(input_scaled)
    print("Input columns at prediction:", input_df.columns.tolist())

    # Confidence as a pseudo-probability
    confidence = round(1 / (1 + np.exp(-pred_prob[0])) * 100, 2)
    prediction = "Diabetic" if pred_class == 1 else "Non-Diabetic"
    return prediction, confidence

# Streamlit UI
st.set_page_config(page_title="Diabetes Prediction App", layout="centered")
st.title("🚑 Diabetes Prediction App")
st.write("Enter patient health data to check for diabetes.")

sample = st.selectbox("Select Sample Patient:", ["Healthy Person", "Diabetic Person", "Custom"])

if sample != "Custom":
    if sample == "Healthy Person":
        input_data = {"Glucose": 90, "BMI": 22, "Age": 25, "Insulin": 80, "SkinThickness": 20}
    else:
        input_data = {"Glucose": 160, "BMI": 35, "Age": 50, "Insulin": 150, "SkinThickness": 35}
else:
    input_data = {}
    input_data['Glucose'] = st.slider("Glucose Level", 0, 200, 100, 1)
    input_data['BMI'] = st.slider("BMI", 10.0, 50.0, 25.0, 0.1)
    input_data['Age'] = st.slider("Age", 0, 100, 30, 1)
    input_data['Insulin'] = st.slider("Insulin Level", 0, 300, 100, 1)
    input_data['SkinThickness'] = st.slider("Skin Thickness", 0, 100, 25, 1)

if st.button("Predict"):
    prediction, confidence = predict_diabetes(input_data)

    st.subheader("Prediction Results")
    st.write(f"**Result:** {prediction}")
    st.write(f"**Prediction Confidence:** {confidence}%")

    # Bar chart
    features = ['Glucose', 'BMI', 'Age', 'Insulin', 'SkinThickness']
    values = [input_data[f] for f in features]
    fig, ax = plt.subplots()
    ax.bar(features, values, color=['#FF6F61','#6B5B95','#88B04B','#F7CAC9','#92A8D1'])
    ax.set_ylabel("Values")
    st.pyplot(fig)

    # CSV download
    df_download = pd.DataFrame([input_data])
    df_download['Prediction'] = prediction
    df_download['Confidence'] = confidence
    csv = df_download.to_csv(index=False).encode()
    st.download_button("📥 Download Prediction (CSV)", csv, "diabetes_prediction.csv", "text/csv")
```
